```python
import os

# 1. 수정이 필요한 라벨들이 들어있는 폴더 (절대 경로 추천)
label_path = r'C:/Code/PythonProject/datasets/labels/train'

# 2. e_stop 이미지 파일명 리스트가 있다면 활용하거나,
# 특정 패턴(예: 'e_stop_'으로 시작하는 파일)이 있다면 조건문을 씁니다.
files = [f for f in os.listdir(label_path) if f.endswith('.txt')]

change_count = 0

for filename in files:
    # 예: 파일명에 'e_stop'이 포함된 라벨 파일만 수정
    if 'e_stop' in filename:
        file_path = os.path.join(label_path, filename)

        with open(file_path, 'r') as f:
            lines = f.readlines()

        new_lines = []
        for line in lines:
            parts = line.split()
            if len(parts) > 0 and parts[0] == '1': # 만약 첫 번째 숫자가 1이라면
                parts[0] = '0' # 0으로 교체
                new_lines.append(" ".join(parts) + "\n")
                change_count += 1
            else:
                new_lines.append(line)

        # 파일 다시 저장
        with open(file_path, 'w') as f:
            f.writelines(new_lines)

print(f"✨ 수정 완료: 총 {change_count}개의 라벨 번호를 1에서 0으로 변경했습니다.")
```
