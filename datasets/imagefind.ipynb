```python
import cv2
import os

def extract_frames(video_path, output_dir):
    # 저장할 폴더 생성
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # 영상 불러오기
    cap = cv2.VideoCapture(video_path)
    count = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # 이미지 파일 저장 (frame_0001.jpg 형식)
        file_path = os.path.join(output_dir, f"left_arrow_{count:04d}.jpg")
        cv2.imwrite(file_path, frame)

        count += 1
        if count % 100 == 0:
            print(f"진행 상황: {count}장 추출 완료")

    cap.release()
    print(f"✅ 총 {count}장의 이미지가 '{output_dir}' 폴더에 저장되었습니다.")

if __name__ == "__main__":
    # 영상 파일 이름이 정확한지 확인하세요!
    video_name = 'left_arrow_video.mp4'

    # 추출된 이미지가 저장될 폴더 이름
    output_folder = 'left_images'

    # 함수 실행
    extract_frames(video_name, output_folder)
```
