```python
import os

# 라벨 파일들이 들어있는 폴더 경로 (팀장님 경로에 맞게 수정)
label_path = r'C:/Code/PythonProject/datasets/haci_labels'

# 폴더 내 모든 파일 확인
files = os.listdir(label_path)

for filename in files:
    # 해시값과 실제 이름 사이의 구분자인 '-'가 있는지 확인
    if '-' in filename:
        # '-'를 기준으로 나누고 가장 마지막 부분(원본 파일명)만 선택
        new_name = filename.split('-', 1)[1]

        old_file = os.path.join(label_path, filename)
        new_file = os.path.join(label_path, new_name)

        # 파일 이름 변경
        try:
            os.rename(old_file, new_file)
            print(f'변경 완료: {filename} -> {new_name}')
        except FileExistsError:
            print(f'이미 파일이 존재함: {new_name}')

print("모든 작업이 완료되었습니다.")
```

```python
import os
import shutil

# 1. 경로 설정 (팀장님 현재 구조 반영)
# 원본 이미지들이 모여있는 곳
#src_images_dir = r'/datasets/images/e_stop_images'
# 이미 옮겨두신 라벨 파일들이 있는 곳
#train_labels_dir = r'/datasets/labels/train_e_stop'
# 라벨과 짝이 맞는 이미지가 들어갈 곳
#train_images_dir = r'/datasets/images/train_e_stop'

# 원본 이미지들이 모여있는 곳
src_images_dir = r'C:/Code/PythonProject/datasets/images/train_images'
# 이미 해시를 제거하고 옮겨두신 라벨 파일들이 있는 곳
train_labels_dir = r'C:/Code/PythonProject/datasets/labels/train_labels'
# 라벨과 짝이 맞는 이미지가 최종적으로 들어갈 곳
train_images_dir = r'C:/Code/PythonProject/datasets/images/testimages'

# 2. 이미지가 들어갈 폴더가 없으면 생성
os.makedirs(train_images_dir, exist_ok=True)

# 3. labels/train 폴더에 있는 모든 .txt 파일 목록 가져오기
labeled_files = [f for f in os.listdir(train_labels_dir) if f.endswith('.txt')]

move_count = 0
for label_file in labeled_files:
    # 확장자 제외 파일명 추출 (예: left_arrow_0000)
    base_name = os.path.splitext(label_file)[0]
    img_name = base_name + '.jpg'  # 이미지 확장자가 .jpg일 경우

    source_img_path = os.path.join(src_images_dir, img_name)
    target_img_path = os.path.join(train_images_dir, img_name)

    # 원본 폴더에 해당 이미지가 있으면 '이동' 시킴
    if os.path.exists(source_img_path):
        shutil.move(source_img_path, target_img_path)
        move_count += 1

print(f"✅ 이동 완료: 총 {move_count}장의 이미지를 'images/train'으로 옮겼습니다.")
```
